### 15.3.1 创建RandomWalk()类

```python
import matplotlib.pyplot as plt
from random import choice

class RandomWalk():
    """一个生成随机漫步数据的类"""
    
    def __init__(self, num_points=5000):
        """初始化随机漫步的属性"""
        self.num_points = num_points
        self.x_values = [0]  # 所有随机漫步都始于(0, 0)
        self.y_values = [0]

    def fill_walk(self):
        """计算随机漫步包含的所有点"""
        # 不断漫步，直到列表达到指定的长度
        while len(self.x_values) < self.num_points:
            # 决定前进方向以及沿这个方向前进的距离，不包括0以避免原地踏步
            x_direction = choice([-1, 1])
            x_distance = choice([1, 2, 3, 4])
            x_step = x_direction * x_distance
            
            y_direction = choice([-1, 1])
            y_distance = choice([1, 2, 3, 4])
            y_step = y_direction * y_distance
            
            # 计算下一个点的x和y值
            next_x = self.x_values[-1] + x_step
            next_y = self.y_values[-1] + y_step
            
            self.x_values.append(next_x)
            self.y_values.append(next_y)

# 使用matplotlib绘制随机漫步图
def draw_random_walk(rw):
    plt.figure(figsize=(10, 6))
    plt.scatter(rw.x_values, rw.y_values, c=range(len(rw.x_values)), cmap='viridis', edgecolors='none', s=15)
    plt.colorbar(label='Step Number')
    plt.title('Random Walk Visualization')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    rw = RandomWalk()
    rw.fill_walk()
    draw_random_walk(rw)
```

### 15.4.5 分析结果 

```python
import random
# 定义一个骰子类（这里假设die类已正确定义并包含了roll方法来返回随机一面的结果）
class Die:
    def __init__(self, num_sides=6):
        self.num_sides = num_sides

    def roll(self):
        return random.randint(1, self.num_sides)

# 初始化骰子
die = Die()

# 模拟掷骰子1000次，并收集结果
results = [die.roll() for _ in range(1000)]

# 分析结果，计算每个点数出现的频率
frequencies = [results.count(value) for value in range(1, die.num_sides + 1)]

# 打印每个面的频率
print(frequencies)
```
